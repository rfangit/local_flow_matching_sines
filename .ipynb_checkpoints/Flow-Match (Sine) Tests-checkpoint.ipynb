# Training Data

```python
def create_phase_dataset(phases, num_points=1000, points_per_cycle=40, device = 'cpu'):
    """
    Takes in a list of phases, the width of each sine wave, and the frequency
    Create a tensor of shape [B, N] where B is number of phases, N is num_points
    """
    dataset = torch.stack([
        generate_sine_wave(num_points, points_per_cycle, phase_offset=phase, device = device)
        for phase in phases
    ])
    return dataset

# Parameters with phase offset for demonstration
num_points = 10
points_per_cycle = 10
# Example usage:
phases = [0]#, math.pi/4, math.pi/2, 3*math.pi/4]  # Example phase offsets
dataset = create_phase_dataset(phases, num_points, points_per_cycle, device = device)

print(dataset.shape)  # [B, 1000]
```

# Locality Dataset Testing

```python
def window_dataset(dataset, window_size, stride=1):
    """
    Apply windowing to each signal in the dataset and stack results
    
    Args:
        dataset: Tensor of shape [B, N]
        window_size: Size of each window
        stride: Stride between windows
        
    Returns:
        Tensor of shape [B * num_windows, window_size]
    """
    # Apply unfolding to each batch element
    windows = [signal.unfold(0, window_size, stride) for signal in dataset]
    
    # Stack all windows across all batches
    return torch.cat(windows, dim=0)
    
window_size = 2

# Create dataset and apply windowing
windowed_data = window_dataset(dataset, window_size, stride = 1)

print("Original dataset:", dataset)  # [B, 1000]
print("Windowed data:", windowed_data)  # [B * num_windows, window_size]

print("Original dataset shape:", dataset.shape)  # [B, 1000]
print("Windowed data shape:", windowed_data.shape)  # [B * num_windows, window_size]
```

# Local Sine Wave Class

```python
class Local_Sine_Wave_Denoiser:
    def __init__(self, original_data: torch.Tensor, sigma_i: float, device ):
        """
            original_data: Tensor of shape [B, N] where B is batch, N is points
        """
        self.original_data = original_data.clone()
        self.sigma_i = sigma_i
        self.device = device

    def set_time(self, new_time: float) -> None:
        """Set a new time value"""
        self.current_time = new_time
    
    def set_local_windows(self, window_size: int, stride = None) -> torch.Tensor:
        """
        Generates windowed view of the current data
            window_size: Size of each window
            stride: Step between windows (default 1)
        Returns:
            Windowed tensor of shape [B * num_windows, window_size], num_windows = (B - window_size) / stride
        """
        if stride is None:
            stride = 1
        # Apply unfolding to each batch element
        windows = [signal.unfold(0, window_size, stride) 
                  for signal in self.original_data]
        
        self.local_data = torch.cat(windows, dim=0).to(self.device)
        return self.local_data

    def match_to_local(self, current_pts_windowed) -> torch.Tensor:
        #assume that target data has been appropriately windowed via some function
        #and we just use the existing function with this
        
        # Create time tensor matching the batch dimension of input
        M = current_pts_windowed.shape[0]  # Number of windows
        time_tensor = torch.full((M,), self.current_time, 
                               device=current_pts_windowed.device)
        #print("\n=== Device Check ===")
        #print(f"current_pts_windowed: {current_pts_windowed.device} (shape: {current_pts_windowed.shape})")
        #print(f"self.local_data: {self.local_data.device} (shape: {self.local_data.shape})")
        
        # Create time tensor and check its device
        M = current_pts_windowed.shape[0]
        time_tensor = torch.full((M,), self.current_time, device=self.device)
        #print(f"time_tensor: {time_tensor.device} (shape: {time_tensor.shape})")
        #print(f"self.sigma_i: {type(self.sigma_i)} (scalar value: {self.sigma_i})")
        #print("===================\n")
        # Compute velocity using properly shaped time tensor
        velocity = compute_linear_velocity_batch_time(
            current_pts_windowed, 
            self.local_data, 
            time_tensor,  # Now passing tensor instead of float
            self.sigma_i
        )
        return velocity
```

# Full Code

We put our parts together. First, we generate the dataset of a variety of sine waves, then we generate batches of noise. We load our dataset into the local sine wave class.

Then we iteratively denoise. At each time step, we pass the current time into the local sine wave, and the current locality size. Then we pass our current batches into our sliding window function along with the local sine wave.match_to_local. This will allow us to calculate the velocity that should act on our current data. We then numerically integrate this.

```python
# Dataset generation
num_points = 100 #points in a sample
points_per_cycle = 20
phases = [0]#, math.pi/4, math.pi/2, 3*math.pi/4]  # Example phase offsets
dataset = create_phase_dataset(phases, num_points, points_per_cycle, device = device)

# Noise generation
# We work with one noise sample.
noise_sample = torch.randn(size=(num_points, ), device=device)
```

```python
# Load our dataset into an instance of local_sine_wave
local_denoiser_dataset = Local_Sine_Wave_Denoiser(dataset, sigma_i = 1.0, device = device)
num_steps = 5
dt = 1.0 / num_steps
locality_sizes = torch.full((num_steps,), 100, device=device) #for each step, use a local window of 100 points.
```

Now we iterate. At each iteration, configure the local sine wave with the time and locality sizes.

```python
## Holds our integration.
trajectory = torch.zeros(num_steps + 1, *noise_sample.shape, device=noise_sample.device)
trajectory[0] = noise_sample.clone()
current_points = noise_sample.clone()
#print ("Example noise ", current_points)
#print ("The dataset is ", dataset)

# The idea of this code is as follows
# we configure the denoising dataset to match the current time and locality
# we break up example noise into a batch of windows with the right locality size
# we send this into compute local velocity batch, and obtain the velocities for each batch
# we take these local velocities and compute the net velocity
current_time = 0
for step in range(1, num_steps + 1):
    # Get current time and locality size, configure denoiser
    current_locality = locality_sizes[step - 1]  # convert to Python int
    local_denoiser_dataset.set_time(current_time)
    local_windows = local_denoiser_dataset.set_local_windows(current_locality)
    #print ("The local windows are " , local_windows)

    windows = current_points.unfold(0, current_locality, 1)
    #print ("The unfolded windows with locality ", current_locality, " are ", windows)
    #print ("The number of windows made is ", len(windows[0]))
    L = len(current_points)

    local_velocities = local_denoiser_dataset.match_to_local(windows)
    #print ("local velocities ", local_velocities)
    
    output = torch.zeros(L, device=device)
    count = torch.zeros(L, device=device)
    
    for i in range(windows.shape[0]):
        start = i
        end = i + current_locality
        output[start:end] += local_velocities[i]
        count[start:end] += 1
    #print ("The number of velocity contributions to each point is ", count)
    net_velocity = output/count
    #print ("The net velocity adjustment is ", net_velocity)
    #print ("To the original point ", current_points)

    current_points = current_points + net_velocity * dt
    current_time = current_time + dt
    trajectory[step] = current_points.clone()
    
    
```

```python
from matplotlib.animation import FuncAnimation

def denoising_animation(trajectory, dataset=None, interval=100):
    """Creates an animation of the denoising process.
    
    Args:
        trajectory: 2D tensor of shape (time_steps, signal_length)
        dataset: Optional target sine wave for comparison (1D tensor)
        interval: Delay between frames in milliseconds
    
    Returns:
        matplotlib.animation.FuncAnimation object
    """
    # Convert to numpy if needed
    if isinstance(trajectory, torch.Tensor):
        trajectory = trajectory.detach().cpu().numpy()
    if dataset is not None and isinstance(dataset, torch.Tensor):
        dataset = dataset.detach().cpu().numpy()
    
    fig, ax = plt.subplots(figsize=(10, 5))
    
    # Initialize lines
    line_noise, = ax.plot([], [], 'b-', label='Denoising Signal', alpha=0.8)
    if dataset is not None:
        line_target, = ax.plot(dataset[0], 'r--', label='Target Sine Wave', alpha=0.5)
    
    # Set up plot
    ax.set_xlim(0, trajectory.shape[1])
    ax.set_ylim(min(trajectory.min(), -1.5), max(trajectory.max(), 1.5))
    ax.grid(True, alpha=0.3)
    ax.set_xlabel('Position')
    ax.set_ylabel('Value')
    ax.set_title('Denoising Process Over Time')
    ax.legend()
    
    # Add text annotation for time step
    time_text = ax.text(0.02, 0.95, '', transform=ax.transAxes)
    
    def update(frame):
        line_noise.set_data(np.arange(trajectory.shape[1]), trajectory[frame])
        time_text.set_text(f'Step: {frame}/{len(trajectory)-1}\nTime: {frame/(len(trajectory)-1):.2f}')
        return line_noise, time_text
    
    anim = FuncAnimation(fig, update, frames=len(trajectory),
                       interval=interval, blit=True)
    plt.close()
    return anim
```

```python
# After your denoising loop...
anim = denoising_animation(trajectory, dataset=dataset)

# To display in Jupyter notebook:
from IPython.display import HTML
HTML(anim.to_jshtml())
```
